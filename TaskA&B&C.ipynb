## Make all the words in "reviewtext" column lowercase.

```python
comments=list(map(lambda x: x.lower(), df.iloc[:,1]))
comments
```

## Import the packages for cleaning data.

```python
from nltk.corpus import stopwords
from nltk.corpus import brown
from nltk.tokenize import word_tokenize,sent_tokenize
from string import punctuation, digits
from collections import Counter
stopwords_set = set(stopwords.words('english'))
brown_set = set(brown.words())
characterstoclean = r'?!,:,/\"-+=@#$%^&*()><{}[]|' + r"'"
punc = punctuation
digi = '[0-9]'
```

## Clean the data

```python
word_list = []
for comment in comments:
    for word in word_tokenize(comment):
         if word in brown_set  and word not in stopwords_set and word not in characterstoclean and word not in punc and word not in digi :
                word_list.append( word )
word_list
```

## Find the word frequency

```python
cnt =Counter(word_list)
cnt.most_common(1000)
```

## Conduct word replacement based on the word in the word frequency result.

```python
food = ['taste','quality','dishes','flavors','ingredients','bowl','selection','variety','choice','meals','serving','choices','indoor','bowls']
service = ['speed', 'waited','friendly','fast','wait','manager','waiting','serve','line','outdoor','bartender','space','environment','bathroom']
price = ['money','priced','charge','charged','paid','paying','dollars'] # bucks
location = ['nearby','downtown','park','parking','campus','located']

new_comments = []
new = str()
for comment in comments:
    for word in word_tokenize(comment):
        if word in service:           
            comment = comment.replace(word,'service')
        elif word in food:
            comment = comment.replace(word,'food')
        elif word in price:
            comment = comment.replace(word, 'price')
        elif word in location:
            comment = comment.replace(word, 'location')
        else:
            comment = comment           
        
    new_comments.append(comment)
print(len(new_comments))
```

## Replace the original values in "reviewtext" column with `new_comments`.

```python
df['reviewtext'] = new_comments
df
```
